```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
X = data.iloc[:,0]
```

```python
Y = data.iloc[:,1]
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
m = 0
c = 0
L = 0.0001
epochs = 1000
n = float(len(X))

for i in range(epochs):
    Y_pred = m*X + c
    D_m = (-2/n)*sum(X * (Y - Y_pred))
    D_c = (-2/n)*sum(Y - Y_pred)
    m = m - L * D_m
    c = m - L * D_c

print(c,m)
```
